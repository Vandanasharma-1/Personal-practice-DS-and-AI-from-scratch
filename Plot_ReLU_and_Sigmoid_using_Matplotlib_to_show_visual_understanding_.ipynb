```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    """Sigmoid activation function."""
    return 1 / (1 + np.exp(-x))

def relu(x):
    """ReLU (Rectified Linear Unit) activation function."""
    return np.maximum(0,x)
```

```python
x=np.linspace(-5, 5, 100)
```

```python
x
```

```python
y_sigmoid=sigmoid(x)
```

```python
y_sigmoid
```

```python
y_relu=relu(x)
y_relu
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
plt.plot(x, y_sigmoid, label='Sigmoid(x)', color='blue')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('Sigmoid(x)')
plt.grid(True)
plt.axhline(0.4, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.legend()

# Plot ReLU
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
plt.plot(x, y_relu, label='ReLU(x)', color='red')
plt.title('ReLU (Rectified Linear Unit) Activation Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.7)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.7)
plt.legend()

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
# plt.savefig('activation_functions.png') # Uncomment to save the plot as an image
plt.show()
```
